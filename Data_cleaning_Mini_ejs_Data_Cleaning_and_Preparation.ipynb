1) Renombren el siguiente dataframe para que las columnas se llamen "primera", "segunda" y "tercera"

```python
import numpy as np
import pandas as pd


df = pd.DataFrame(np.random.randn(5,3))
df.iloc[:3, 2] = np.nan
df.iloc[2:, 1] = np.nan
df.iloc[2, 0] = np.nan
df
```

```python
df.columns=['Primera', 'Segunda', 'Tercera']
df
#df.rename(columns={0:'Primera', 1:'Segunda', 2:'Tercera'})## otra forma de hacerlo
#df
```

```python
df.dropna(axis=0, how= 'any')##hay filas y columnas que tienen nan por eso borra todo tanto como en filas como en columnas
```

```python
df.dropna(axis=1, how='any')
```
