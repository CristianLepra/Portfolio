## Normalización

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
del df['Unnamed: 0']
```

```python
df
```

```python
df_train = df.iloc[:25]
df_test = df.iloc[25:]
X_train = df_train[['Salary']]
X_test = df_test[['Salary']]
display (X_train)
display (X_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Salary']], df[['YearsExperience']], test_size=5, random_state=42)
```

```python
X_train, X_test, y_train, y_test
```

```python
# Limpieza de datos: normalización.
normalizer = preprocessing.MinMaxScaler()
X_train_norm = normalizer.fit_transform(X_train) 
print(X_train_norm)
```
