```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('WorldCups.csv')
df
```

```python
df.info()
```

```python
#Organizadores - Pais Sede x Año
df[['Year','Country']]
```

```python
df2 = df[['Year','GoalsScored']]
df2
```

```python
#Goles por Mundial
df2.plot.bar(x='Year',y='GoalsScored')

plt.xlabel('Años')
plt.ylabel('Cantidad de goles')
plt.show()
```

```python
#Goles por Mundial
#una forma
fig = plt.figure()
ax = fig.subplots()
ax.bar(df['Year'], df['GoalsScored'])
plt.show()
```

```python
df
```

```python
#Cantidad de Equipos por Mundial?
fig = plt.figure(figsize=(8, 8))
ax = fig.subplots()

ax.plot(df['Year'], df['QualifiedTeams'], marker = 'o', color="tab:purple")

plt.show()
```

```python
#Cantidad de goles por Mundial?
fig = plt.figure(figsize=(8, 8))
ax = fig.subplots()
x = df['Year']
y = df['GoalsScored']
ax.plot(x, y, marker = 'o', color="tab:purple", label = df['Year'])

#annotations=["Punto 1", "Punto 2", "Punto 3", "Punto 4", "Punto 5"]
# plt.scatter(x, y, label="Valores")
# plt.xlabel("X")
# plt.ylabel("Valores")
for i, label in enumerate(df['Country']):
    plt.annotate(df['Country'], (x[i] + 0.1, y[i]))
    
plt.show()
```

```python
#Audiencia?
fig = plt.figure(figsize=(8, 8))
ax = fig.subplots()

ax.plot(df['Year'], df["Attendance"].str.replace(".", "", regex=True).astype(int), marker = 'o', color="tab:purple", label = df['Year'])

plt.show()
```

```python
x=df['Year']
y=df.Attendance.str.replace(".", "", regex=True).astype(int)
plt.scatter(x, y, label=df['Country'])
plt.xlabel("Años")
plt.ylabel("Audiencia")
for i, label in enumerate(df['Country']):
    plt.annotate(label, (x[i] + 0.1, y[i]))
    
plt.show()
```

```python
#Top Campeones
winners = df.groupby('Winner')['Country'].count().sort_values(ascending=False)
winners
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.subplots()
ax.bar(winners.index, winners.values)

labels = ax.set_xticklabels(labels = winners.index, rotation=30, fontsize='small')
plt.show()
```

```python
#Top Sub campeones
df.groupby('Runners-Up')['Country'].count().sort_values(ascending=False)
```

```python
#Top terceros
df.groupby('Third')['Country'].count().sort_values(ascending=False)
```
