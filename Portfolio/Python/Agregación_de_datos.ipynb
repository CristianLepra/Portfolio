- 11) Chequeen si hay valores faltantes. Si hay, rellenenlos con la mediana agrupando por tienda, categoría y subcategoría </br>
Ps: usen el método apply y una función definida por ustedes

```python
df
```

```python
# sin hacerlo con funcion ni apply
#Calcular la mediana por grupo
medianas = df.groupby(['Store', 'Category', 'Subcategory'])[['Dollars', 'Units']].transform('median')

# Llenar valores faltantes con las medianas por grupo
df = df.fillna(medianas)
df
```

```python
def relleno(x):
  x['Dollars'].fillna(x['Dollars'].median(), inplace=True)
  x['Units'].fillna(x['Units'].median(), inplace=True)
  return(x)
```

```python
df = df.groupby(['Store','Category','Subcategory']).apply(relleno)
df
```
